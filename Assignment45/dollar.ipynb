```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from lls import LinearLeastSquare
import datetime as dt
import matplotlib.dates as mdates
import re
from lls import LinearLeastSquare
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('Dataset\Dollar_Rial_Price_Dataset.csv')

data
```

```python
data.dtypes
```

```python
data['Year_month'] = pd.to_numeric(pd.to_datetime(data['Date']).dt.strftime("%Y%m%d"))

data
```

```python
data['Close'] = data['Close'].apply(lambda x: re.sub(',', '', x))
data["Close"] = pd.to_numeric(data["Close"] , errors='coerce')

data['Low'] = data['Low'].apply(lambda x: re.sub(',', '', x))
data["Low"] = pd.to_numeric(data["Low"] , errors='coerce')

data['High'] = data['High'].apply(lambda x: re.sub(',', '', x))
data["High"] = pd.to_numeric(data["High"] , errors='coerce')

data
```

```python
data.dtypes
```

```python
df_ahmadinezhad = data.query('Year_month <= 20130822')

df_ahmadinezhad
```

```python
df_rouhani = data.query('20130822< Year_month <= 20210822')

df_rouhani
```

```python
df_raeeisi = data.query('20210822 < Year_month')

df_raeeisi
```

```python
print('Max price of dollar in Ahmadinejad presidency:',df_ahmadinezhad['High'].max())
print('Max price of dollar in Rouhani presidency:',df_rouhani['High'].max())
print('Max price of dollar in Raeisi presidency:',df_raeeisi['High'].max())
```

```python
print('Min price of dollar in Ahmadinejad presidency:',df_ahmadinezhad['Low'].min())
print('Min price of dollar in Rouhani presidency:',df_rouhani['Low'].min())
print('Min price of dollar in Raeisi presidency:',df_raeeisi['Low'].min())
```

Train Test Split Dataset

Ahmadinejad presidency

```python
X = np.array(df_ahmadinezhad['Year_month'])
Y = np.array(df_ahmadinezhad['Close'])

X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

print(X.shape)
print(Y.shape)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y)

print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
lls = LinearLeastSquare()
lls.fit(X_train,Y_train)
Y_pred = lls.predict(X_test)
```

```python
MAE = lls.evaluate(X_test, Y_test, metric="mae")

print('Mean Absolute Error Ahmadinejad presidency:',MAE)
```
