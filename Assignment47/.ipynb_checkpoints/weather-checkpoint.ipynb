```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

```python
df = pd.read_csv('weatherHistory.csv', encoding='utf-8-sig')

df
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['Formatted Date'] = df['Formatted Date'].dt.date
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['dayofyear'] = df['Formatted Date'].dt.dayofyear
df['year'] = df['Formatted Date'].dt.year

df.head()
```

```python
df_day_mean = df.groupby(['dayofyear','year'])['Temperature (C)'].mean().reset_index()

df_day_mean
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df_day_mean['dayofyear'],df_day_mean['Temperature (C)'],marker='.')
plt.xlabel('days of year')
plt.ylabel('Temperature')
plt.show()
```

```python
class Perceptron:
    def __init__(self,learning_rate,input_length,epochs):
        self.lr = learning_rate
        self.w = np.random.rand(input_length)
        self.b = np.random.rand(1)
        self.epochs = epochs

    def activation(self,x,function='relu'):
        if function == 'sigmoid':
            return 1/(1+np.exp(-x))
        elif function == 'relu':
            return np.maximum(0,x)
        elif function == 'tanh':
            return np.tanh(x)
        elif function == 'linear':
            return x
        else:
            return Exception('Unknown activation function')        


    def fit(self,X_train,Y_train,X_test,Y_test):
        losses = []
        accuracies= []
        losses_test = []
        accuracies_test = []
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))


        for epoch in tqdm(range(self.epochs)):
            for x,y in zip(X_train,Y_train):
                # forwarding
                y_pred = self.w @ x + self.b
                y_pred = self.activation(y_pred,'sigmoid')

                #back propagation
                error = y -y_pred

                # update
                self.w = self.w + x * error * self.lr
                self.b = self.b + error * self.lr
            loss_train,accuracy_train = self.evaluate(X_train,Y_train)
            loss_test,accuracy_test = self.evaluate(X_test,Y_test)

            losses.append(loss_train)
            accuracies.append(accuracy_train)
            losses_test.append(loss_test)
            accuracies_test.append(accuracy_test)

            ax1.plot(accuracy_train)
            ax1.plot(accuracies_test)
            ax1.set_title("Model Accuracy")
            ax1.set_xlabel("epoch")
            ax1.set_ylabel("Accuracy")
            ax1.legend(['train','test'])


            ax2.plot(loss_train)
            ax2.plot(loss_test)
            ax2.set_title("Model loss")
            ax2.set_xlabel("epoch")
            ax2.set_ylabel("Loss")
            ax2.legend(['train','test'])

            plt.show()

        #np.save('wandb_train.npy', self.W+self.B)


    def predict(self,X_test,Y_test):
        Y_pred = []
        losses = []
        accuracies= []
        for x_test in X_test:
            y_pred = x_test @ self.w + self.b
            y_pred = self.activation(y_pred)
            loss_train,accuracy_train = self.evaluate(X_test,Y_test)
            losses.append(loss_train)
            accuracies.append(accuracy_train)
            Y_pred.append(y_pred)
        return np.array(Y_pred),losses,accuracies

    def calculate_loss(self,X_test,Y_test,metric='mae'):
        Y_pred = self.predict(X_test,Y_test)
        if metric == "mae" :
            return np.sum(np.abs(Y_test - Y_pred))
        elif metric == "mse":
            return np.sum((Y_test - Y_pred) ** 2)
        elif metric == "rmse" :
            return np.sqrt(np.sum((Y_test - Y_pred) ** 2))
        
        else:
            return Exception('Unknown metric')  
        
    def calculate_accuracy(self,X_test,Y_test):
        Y_pred = self.predict(X_test,Y_test)
        Y_pred = np.where(Y_pred > 0.5,1,0)
        accuracy = np.sum(Y_pred == Y_test) / len(Y_test)
        return accuracy
    
    def evaluate(self,X_test,Y_test):
        loss = self.calculate_loss(X_test,Y_test)
        accuracy = self.calculate_accuracy(X_test,Y_test)

        return loss,accuracy
```

```python
X = df_day_mean['dayofyear'].values
Y = df_day_mean['Temperature (C)'].values
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y)

print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
model = Perceptron(learning_rate=0.00001,input_length=X_train.shape[1],epochs=64)
model.fit(X_train,Y_train,X_test,Y_test)
```
